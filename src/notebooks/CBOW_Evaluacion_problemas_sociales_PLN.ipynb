```python
df_nlp['DESC_CLEAN']=df_nlp.apply(lambda x:str(x['DESCRIPCION_LEM']).replace(",","").replace("'","").replace("[","").replace("]","").split() ,axis=1)
```

```python
df_nlp.head()
```

**Eliminando stop words**

```python
stop_words = nltk.corpus.stopwords.words('spanish')
```

```python
stop_words[:10]
```

```python
df_nlp['DESC_SIN_SW']=df_nlp.apply(lambda x:[ i for i in x['DESC_CLEAN'] if i not in stop_words] , axis=1 )
```

```python
df_nlp.head(10)
```

**Creando CBOW con Gensim**

```python
cbow_model=Word2Vec(df_nlp['DESC_SIN_SW'], vector_size=5,seed=123)
```

```python
cbow_model_vectors =cbow_model.wv
result= cbow_model_vectors.similar_by_word("sistema")
print("Parecidas a la palabra", result[:5])
```

```python
x=cbow_model.wv[cbow_model.wv.index_to_key]
```

```python
tsne=TSNE(n_components=2, n_iter=250)
X2d=tsne.fit_transform(x)
```

```python
cbow_df=pd.DataFrame(X2d, columns=['x','y'])
cbow_df['token']=cbow_model.wv.key_to_index
cbow_df.head()
```

**Grafico**

```python
output_notebook()
p= figure(plot_width =800, plot_height=800)
_=p.text(x=cbow_df.x, y=cbow_df.y, text= cbow_df.token)
show(p)
```

**Preparar Embedding**

```python
words = set(cbow_model.wv.index_to_key )
```

```python
df_nlp['CBOW_VECT'] = np.array([np.array([cbow_model.wv[i] for i in ls if i in words])
                         for ls in df_nlp['DESC_SIN_SW']])
```

```python
df_nlp.head()
```

```python
text_vect_avg = []
for v in df_nlp['CBOW_VECT']:
    if v.size:
        text_vect_avg.append(v.mean(axis=0))
    else:
        text_vect_avg.append(np.zeros(5, dtype=float)) 
        
        
df_nlp['CBOW_VECT_AVG'] = text_vect_avg
```

**Creando DF Final**

```python
df_nlp["TIPO"].value_counts()
```

```python
df_cbow = pd.DataFrame(text_vect_avg)
df_cbow.head(2)
```

```python
df_cbow.columns = ['variable_' + str(i+1) for i in range(0, df_cbow.shape[1])]
df_cbow.head(2)
```

**DF Final**

```python
df_cbow_final = pd.concat([df_nlp['TIPO'], df_cbow], axis=1, sort=False)
df_cbow_final.head(100)
```
